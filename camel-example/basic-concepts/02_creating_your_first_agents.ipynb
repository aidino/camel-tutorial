## Philosophical Bits - Những suy ngẫm triết học

Lớp `ChatAgent()` là nền tảng của CAMEL 🐫. Chúng tôi thiết kế agent của mình với tinh thần nhằm trả lời câu hỏi sau đây:
- Liệu chúng ta có thể thiết kế một tác nhân giao tiếp tự chủ có khả năng dẫn dắt cuộc hội thoại hướng đến việc hoàn thành nhiệm vụ với sự giám sát tối thiểu từ con người?
- Trong cách triển khai hiện tại của chúng tôi, chúng tôi xem xét các tác nhân với những đặc điểm chính sau:
    - **Role**: cùng với mục tiêu và đặc tả nội dung, yếu tố này thiết lập trạng thái ban đầu của một tác nhân, hướng dẫn tác nhân thực hiện các hành động trong quá trình tương tác tuần tự.
    - **Mô hình Ngôn ngữ Lớn (LLMs)**: mỗi tác nhân sử dụng một Large Language Model để nâng cao khả năng nhận thức. LLM cho phép hiểu và tạo ngôn ngữ tự nhiên, giúp các tác nhân diễn giải chỉ dẫn, tạo phản hồi và tham gia vào các cuộc đối thoại phức tạp.
    - **Bộ nhớ**: bộ nhớ trong ngữ cảnh và bộ nhớ bên ngoài, cho phép tác nhân suy luận và học tập theo cách tiếp cận thực tế hơn.
    - **Công cụ**: một tập hợp các chức năng mà các tác nhân của chúng tôi có thể sử dụng để tương tác với thế giới bên ngoài; về cơ bản, điều này mang lại hiện thân cho các tác nhân của chúng tôi.
    - **Giao tiếp**: khung làm việc của chúng tôi cho phép giao tiếp linh hoạt và có khả năng mở rộng giữa các tác nhân. Đây là nền tảng cho câu hỏi nghiên cứu then chốt.
    - **Suy luận:** chúng tôi sẽ trang bị cho các tác nhân những khả năng lập kế hoạch và học tập phần thưởng (phản biện) khác nhau, cho phép chúng tối ưu hóa việc hoàn thành nhiệm vụ theo cách tiếp cận có định hướng hơn.

### Quick Start 

#### 🕹 Step 1: Define the Role

```python
sys_msg = 'You are a curious stone wondering about the universe.'
```

#### 🕹 Step 2: Set up the Model

Sử dụng `ModelFactory` để thiết lập mô hình backend cho agent, để biết thêm chi tiết về các cài đặt mô hình, vui lòng truy cập tài liệu mô hình của chúng tôi.


```python
from camel.models import ModelFactory
from camel.types import ModelPlatformType, ModelType
from camel.configs import ChatGPTConfig

# Define the model, here in this case we use gpt-4o-mini
model = ModelFactory.create(
    model_platform=ModelPlatformType.OPENAI,
    model_type=ModelType.GPT_4O_MINI,
    model_config_dict=ChatGPTConfig().as_dict(), # [Optional] the config for model
)
```

Set `ChatAgent`

```python
from camel.agents import ChatAgent
agent = ChatAgent(
    system_message=sys_msg,
    model=model,
    message_window_size=10, # [Optional] the length for chat memory
    output_language="Vietnamese"
    )
```

#### 🕹 Step 3: Interact with the Agent with `.step()`

```python
# Define a user message
usr_msg = 'what is information in your mind?'

# Sending the message to the agent
response = agent.step(usr_msg)

# Check the response (just for illustrative purpose)
print(response.msgs[0].content)
```

### Advanced Features

#### 🔧 Tool Usage  

```python
# Import the necessary tools
from camel.toolkits import MathToolkit, SearchToolkit

SearchToolkit().get_tools()
```

```python


# Initialize the agent with list of tools
agent = ChatAgent(
    system_message=sys_msg,
    tools = [
        *MathToolkit().get_tools(),
        SearchToolkit().search_serper,
    ],
    output_language="Vietnamese"
    )
# Let agent step the message
response = agent.step("What is CAMEL AI?")

# Check tool calling
print(response.info['tool_calls'])

# Get response content
print(response.msgs[0].content)
```

#### 🧠 Memory 

```python
agent.memory.get_context()
```

Bạn có thể cập nhật/thay đổi bộ nhớ của agent bằng bất kỳ tin nhắn nào được cung cấp từ bên ngoài theo định dạng `BaseMessage` ; ví dụ: sử dụng một tin nhắn người dùng mới:


```python
from camel.messages import BaseMessage

new_user_msg = BaseMessage.make_user_message(
    role_name="CAMEL User",
    content="This is a new user message would add to agent memory",
)

# Update the memory
agent.record_message(new_user_msg)
```
